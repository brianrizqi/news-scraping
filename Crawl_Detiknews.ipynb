# CNN

```python
cnn = cnn.reset_index(drop=True)
cnn
```

```python
def get_content(e):
  tag_a = e.find('a')
  if 'ADVERTISEMENT' in e.text or 'SCROLL TO CONTINUE WITH CONTENT' in e.text:
    return ''
  if tag_a is not None:
    if tag_a.get('id') == 'idvideocnn':
      return ''
  return e.text
```

```python
headers = {"user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.182 Safari/537.36"}
url = 'https://www.cnnindonesia.com/ekonomi/20240131063721-532-1056475/bansos-tahu-bulat-jokowi-di-masa-kampanye-pilpres-untuk-siapa'
page = requests.get(url,headers=headers)
soup = BeautifulSoup(page.content, 'html.parser')
section = soup.find('div',{'class' : 'flex gap-2 text-base'})
content = ''
if section is not None:
  page_length = len(section.find_all('a'))
  for num in range(1,page_length+1):
    url_page = url + '/'+ str(num)
    page = requests.get(url_page,headers=headers)
    soup = BeautifulSoup(page.content, 'html.parser')
    section = soup.find('div',{'class' : 'detail-text'})
    section = section.extract()
    section = section.find_all('p')
    for e in section:
      content += get_content(e)
else:
  section = soup.find('div',{'class' : 'detail-text'})
  section = section.extract()
  section = section.find_all('p')
  for e in section:
    content += get_content(e)
print(content)
```

```python
def tempo_get_content(e):
  div = e.find('div')
  if div is not None:
    return ''
  return e.text
```

```python
headers = {"user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.182 Safari/537.36"}
url = 'https://bisnis.tempo.co/read/1590564/tinjau-lokasi-kebakaran-mentan-minta-tidak-boleh-ada-perbaikan-kapal-di-dermaga'
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
content = ''
section = soup.find('ul',{'class' : 'pagging'})
if section is None:
  section = soup.find('div',{'class' : 'detail-konten'})
  section = section.extract()
  section = section.find_all('p')
  for e in section:
    if "pilihan editor" in e.text.lower() or "selalu update info terkini" in e.text.lower():
        break
    if "baca juga" in e.text.lower():
        continue
    content += tempo_get_content(e)
  print(content)
else:
  li = section.find_all('li')
  for num_page in range(1,len(li)):
    url_page = url + '?page_num=' + str(num_page)
    page = requests.get(url_page)
    soup = BeautifulSoup(page.content, 'html.parser')
    li_tags = soup.find_all('ul')
    for li in li_tags:
      li.decompose()
    section = soup.find('div',{'class' : 'detail-konten'})
    section = section.find_all('p')
    for e in section:
      if "pilihan editor" in e.text.lower() or "selalu update info terkini" in e.text.lower():
        break
      if "baca juga" in e.text.lower():
        continue
      content += tempo_get_content(e)
  print(content)
```
